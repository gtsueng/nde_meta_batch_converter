# Reverse resourceCatalog converter

This notebook converts the selected `resourceCatalogs/*.json` records back into a multi-sheet workbook that follows the `data/*.xlsx` template.

Notes:
- The workbook keeps the same sheets and columns as the existing template.
- Some JSON properties are richer than the spreadsheet schema. When that happens, the converter writes the closest matching columns and stores unmapped object fields in `related.description` as `EXTRA_JSON`.

### Reverse convert manually generated ResourceCatalogs

The original high priority resource catalogs were curated manually via the DDE and exported via the DDE using the DDE's markup generator with some post-export edits. This means that the manually generated ResourceCatalogs are not available in the RepoMetaCuration google spreadsheet which is useful for batch updating json files whenever there is a schema change.

To ensure that we'll be able to batch update these records, this reverse conversion will be performed once in order to get the records into the RepoMetaCuration google spreadsheet

```python
from pathlib import Path

from reverse_xlsx_batch_converter import convert_resource_catalogs
```

```python
DEFAULT_TARGET_FILES = [
    "ACTG.json",
    "BV-BRC.json",
    "IEDB.json",
    "ITN TrialShare.json",
    "mwccs.json",
    "TBPortals.json",
    "ImmuneSpace.json",
]
```

```python
SCRIPT_PATH = Path.cwd()
INPUT_DIR = SCRIPT_PATH.parent / 'nde-metadata-corrections' / 'metadata_for_DDE' / 'resourceCatalogs'
OUTPUT_DIR = SCRIPT_PATH / 'data'
TARGET = DEFAULT_TARGET_FILES
OUTPUT_PATH = OUTPUT_DIR / 'RepoMetaCuration_reverse.xlsx'

output_path = convert_resource_catalogs(
    json_dir=INPUT_DIR,
    output_path=OUTPUT_PATH,
    target=TARGET,
)
output_path
```

### Reverse convert API-generated ResourceCatalogs

With the intention to develope a repository matcher, we will need to reconcile metadata differences between the sources themselves (i.e. the Repository-level metadata) vs the ResourceCatalog. The ResourceCatalogs are rich and contain many metadata fields desirable for repository searching. In contrast, much of this metadata is unavailable for most ingested repositories.

The api_batch_generator.ipynb was used to generate the base ResourceCatalog json files for each ingested resource; however, there is no easy way to update all of them in the event of a schema change or with any new curation needed for the repository finder/matcher.

This section of the notebook converts the base ResourceCatalog JSON files into the RepoMetaCuration spreadsheet format so that a separate spreadsheet can be maintained for the metadata for ingested repositories

```python
from pathlib import Path
import os
from reverse_xlsx_batch_converter import convert_resource_catalogs
```

```python
SCRIPT_PATH = Path.cwd()
INPUT_DIR = SCRIPT_PATH.parent / 'nde-metadata-corrections' / 'metadata_for_DDE' / 'dataCatalogs'
OUTPUT_DIR = SCRIPT_PATH / 'data'
OUTPUT_PATH = OUTPUT_DIR / 'RepoMetaCuration_ingested_sources.xlsx'
```

```python
potential_targets = os.listdir(INPUT_DIR)
TARGET = potential_targets[2:]
```

```python
output_path = convert_resource_catalogs(
    json_dir=INPUT_DIR,
    output_path=OUTPUT_PATH,
    target=TARGET,
)
output_path
```
